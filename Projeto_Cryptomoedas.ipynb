```python
#Função para converter valores em notação científica para notação decimal
def converter_notacao_cientifica(valor):
    try:
        return "{:.10f}".format(float(valor))
    except:
        return valor
#Colunas a serem convertidas
colunas_para_converter = ['Alto', 'Baixo', 'Aberto', 'Fechado']

for coluna in colunas_para_converter:
    crypto_coins[coluna] = crypto_coins[coluna].apply(lambda x: converter_notacao_cientifica(x))
```

```python
urls_arquivos_csv = [
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_Aave.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_BinanceCoin.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_Bitcoin.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_ChainLink.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_Cosmos.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_EOS.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_Ethereum.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_Polkadot.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_USDCoin.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/ArquivosCrypto/coin_XRP.csv'
]
```

```python
dados_moedas = {}

for url in urls_arquivos_csv:
    # Extrai o nome da moeda do URL
    nome_moeda = url.split('/')[-1].replace('coin_', '').replace('.csv', '')

    # Carrega o arquivo CSV em um DataFrame
    df = pd.read_csv(url)

    # Armazena o DataFrame no dicionário usando o nome da moeda como chave
    dados_moedas[nome_moeda] = df
```

```python
# Plot dos valores de todas as criptomoedas
plt.figure(figsize=(12, 6))
for moeda, df in dados_moedas.items():
    plt.plot(df['Date'], df['Close'], label=moeda)

plt.xlabel('Data')
plt.ylabel('Valor de Fechamento')
plt.title('Valores de Fechamento de Criptomoedas ao Longo do Tempo')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Dicionário para armazenar a maior variação percentual de preço para cada moeda
maior_variacao = {'Moeda': None, 'Data': None, 'Variação Percentual': None}

# Iterar pelas moedas
for moeda, df in dados_moedas.items():
    # Calcular a variação percentual de preço para cada dia
    df['Variação Percentual'] = (df['Close'] / df['Open'] - 1) * 100
    # Encontrar o dia com a maior variação percentual
    dia_maior_variacao = df.loc[df['Variação Percentual'].idxmax()]

    # Se a variação encontrada for maior do que a armazenada, atualizar o dicionário
    if maior_variacao['Variação Percentual'] is None or dia_maior_variacao['Variação Percentual'] > maior_variacao['Variação Percentual']:
        maior_variacao['Moeda'] = moeda
        maior_variacao['Data'] = dia_maior_variacao['Date']
        maior_variacao['Variação Percentual'] = dia_maior_variacao['Variação Percentual']

# Exibir a moeda com a maior variação percentual em um único dia
print(f'A criptomoeda com a maior variação percentual de preço em um único dia durante o período analisado é {maior_variacao["Moeda"]}.')
print(f'Isso ocorreu em {maior_variacao["Data"]} com uma variação de {maior_variacao["Variação Percentual"]:.2f}%.')
```

```python
# Dicionário para armazenar as maiores variações percentuais
maiores_variacoes = {'Moeda': [], 'Variação Percentual': []}

# Iterar pelas moedas
for moeda, df in dados_moedas.items():
    # Encontrar o dia com a maior variação percentual
    dia_maior_variacao = df.loc[df['Variação Percentual'].idxmax()]
    # Adicionar a moeda e a variação ao dicionário
    maiores_variacoes['Moeda'].append(moeda)
    maiores_variacoes['Variação Percentual'].append(dia_maior_variacao['Variação Percentual'])

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(maiores_variacoes['Moeda'], maiores_variacoes['Variação Percentual'])
plt.xlabel('Moeda')
plt.ylabel('Variação Percentual (%)')
plt.title('Maior Variação Percentual de Preço em um Único Dia por Moeda')
plt.xticks(rotation=45)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
