2. **Ethereum (ETH) - Valor Médio $383,91:** O Ethereum é a segunda maior criptomoeda por capitalização de mercado e continua a desempenhar um papel fundamental no mercado devido à sua plataforma de contratos inteligentes e aplicativos descentralizados (DApps).

10. **XRP (XRP) - Valor Médio $0,24:** O XRP é usado principalmente para transferências de dinheiro transfronteiriças, com um valor médio que o torna uma escolha eficiente para soluções de pagamento global.


## Em quais anos houve quedas e valorizações?

Para realizar esta análise as moedas foram separadas em dois quadrantes contendo cinco em cada, medindo a linha temporal desde o surgimento delas até o marco de 2021.

Esta divisão foi realizada, primeiramente, para facilitar a visualização mais limpa das informações, e depois, para enquadrar melhor o período de surgimento/início da operação de cada moeda dentro de uma margem.

A criação dos dois gráficos foram feitas com base nos valores registrados em "Fechamento" disponível na tabela, e para verificar as altas e as quedas do fechamento, foi efetuada a média para indentificar os preços nos quais cada cripto fechou o ano, assim pode-se tirar a base de valorização.

O primeiro gráfico nos mostra o balanço das moedas ChainLink (LINK), Cosmos, EOS, Polkadot e USDCoin.
* Da data de criação/lançamento destas criptos, durante seu tempo de vida, nota-se que 2021 foi o ano de maior valor para ChainLink, Polkadot e Cosmos.
* Para a EOS, seu pico foi no ano de 2018 com tendência de congelamento após o ano de 2019
* USDCoin manteve linearidade na estabilidade do seu preço.

Enquanto as demais criptomoedas variam suas tendências, a USDCoin se mantém constante. Isso ocorre porque ela é uma moeda pareada ao dólar só que funcionando em carteiras digitais e com o sistema blockchain, então, técnicamente, seu valor se estabiliza no preço da moeda norte-americana.

A curva da EOS tem alta logo no início de sua operação e depois, uma queda suave com predição de estagnar. Há alguns fatores que podem dar significado a este movimento. A EOS surgiu com um padrão diferenciado de tecnologia blockchain descentralizada, a fácil operação e utilização de tokens (ativos que possuem valor comercial) da moeda gerou bilhões para sua receita.

O movimento de decaída da EOS se deve a problemas internos em relação ao desenvolvimento contínuo da sua ferramenta base (EOSIO), causando falhas e devido a falta de recursos, causou queda na cripto, até 2021, em que surgiu a ENF (EOS Network Foundation) com nova proposta para levantar a moeda.

As criptomedas ChainLink, Polkadot e Cosmos seguem a tendência de subida do gráfico, quase de maneira simultânea. O movimento da Cosmos pode estar ligado ao seu modo de operação, é uma cripto interoperadore, que permite transferencia entre outras moedas, o que pode ter valorizado e permitido a escalabilidade. Já a ChainLink (LINK) e a Polkadot (DOT) iniciaram como plataformas de transação de criptomoedas para superar 'gargalos' deste mercado, como a de criar pontes  entre criptos e a facilidade para os usuários, antes de iniciarem suas operações com moeda própria, o que em si foi um facilitador e assim que suas moedas foram lançadas houve supervalorização.

O segundo gráfico nos mostra o balanço das moedas Aave (AAVE), Binance Coin (BNB), Bitcoin (BTC), Ethereum (ETH) e XRP (XRP).
* As criptos Bitcoin, Ethereum, Aave e Binance, respectivamente, tiveram suas cotações mais altas em 2021;
* XRP mantém sua linearidade abaixo de $1 dólar, sua variação é imperceptível.

A criptomoeda Aave é recente e já alcançou um pico de sucesso 1 ano após o início de operação. Um ponto que pode justificar a crescente é que a plataforma de investimento (o usuário recebe juros por aplicação) também permite empréstimos em troca de pagar taxas.

O fato da XRP não demonstrar curvas em sua linha pode estar ligado a investigação da SEC (Securities and Exchange Commission, Comissão de Segurança dos bancos de Criptoativos) sobre a suspeita da unidade ter negociado acordos com entidades não registradas na SEC para elencar a XRP no intuito de atingir métricas de volume e poder valorizar a cripto. Esta negociação com mercados secundários podem justificar o impacto na estagnação da margem de preço.

A Binance, inicialmente uma corretora de moedas, começou sua operação em 2017 na China, sendo um ano não tão bom devido as mudanças internas na economia chinesa em que o governo baniu a atividade de criptoativos no país asiático.
Assim como Bitcoin e Ethereum, e outras moedas, teve ascensão no mercado em 2021, e este aumento pode estar relacionado à época em que o mundo estava, fragilizado pelas consequências da Covid-19, vários setores do mercado monetário comum sofrendo com inflações e ofertas de créditos demasiada, e isto pode ter impulsionado seu crescimento devido ao valor neste momento e a busca por estabilidade e segurança em investimentos pelos usuários, além de que a Binance também oferece aos investidores retorno em juros, além da opção de negociar várias moedas dentro de sua plataforma criptofinanceira.

O Bitcoin é a criptomoeda que mais sofre oscilação, no entanto, suas altas sempre são maiores que a queda. Esta ocorrência pode estar dretamente relacionada ao motivo do Bitcoin ser uma moeda limitada mundialmente. Ou seja, ela foi criada com um limite, 21 milhões de moedas, apenas. Por estas vias, ela se torna uma fonte de "ouro" e seu funcionamento entra na questão da oferta e demanda no mercado, por isso, quanto mais moedas são "mineiradas" e entram na carteira de alguém, menos Bitcoins existem fazendo com que se tornem cada vez mais caros.

Ao examinar as curvas da criptomoeda Ethereum, é possível notar com ela também sofre com oscilações, menos que o Bitcoin, mas ainda assim tem suas fases. Como todo o restante do mercado de moedas digitais que se beneficiaram em 2021 com a ascensão do mercado e mais adesão no momento pandêmico, a elevação da Ethereum, em sua particularidade, pode estar relacionada aos NFTs (Tokens Não Fungíveis), que são itens/símbolos exclusivos e não podem ser duplicados. O movimento de NFTs também surgiu em 2021 e ele é negociado em criptomoedas e criado dentro das plataformas exchange (corretoras de criptomoedas), e a maior delas é a Ethereum.
